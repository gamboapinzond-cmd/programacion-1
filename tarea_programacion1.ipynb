```python
def suma(z1, z2):
    return z1 + z2

def multiplicacion(z1, z2):
    return z1 * z2

def raiz_n(z, n):
    return z ** (1/n)

def pedir_complejo():
    while True:
        try:
            real = float(input("Ingrese la parte real: "))
            imag = float(input("Ingrese la parte imaginaria: "))
            return complex(real, imag)
        except ValueError:
            print("Error: Ingrese solo valores numericos.")

while True:
    print("\nMENU DE OPCIONES")
    print("1. Sumar numeros complejos")
    print("2. Multiplicar numeros complejos")
    print("3. Raiz n-esima de un numero complejo")
    print("4. Salir")

    opcion = input("Elija una opcion: ")

    if opcion == '1':
        z1 = pedir_complejo()
        z2 = pedir_complejo()
        print("La suma es:", suma(z1, z2))
    elif opcion == '2':
        z1 = pedir_complejo()
        z2 = pedir_complejo()
        print("La multiplicacion es:", multiplicacion(z1, z2))
    elif opcion == '3':
        z = pedir_complejo()
        while True:
            try:
                n = float(input("Ingrese el valor de n: "))
                if n == 0:
                    print("Error: n no puede ser 0.")
                    continue
                break
            except ValueError:
                print("Error: Ingrese un numero valido.")
        print("La raiz es:", raiz_n(z, n))
    elif opcion == '4':
        print("chao profe")
        break
    else:
        print("Opcion no valida.")
```

```python
def pedir_vector(mensaje):
    while True:
        try:
            entrada = input(mensaje).split()
            return [float(x) for x in entrada]
        except ValueError:
            print("Error: Ingrese solo numeros separados por espacios.")

while True:
    print("\nMENU DE OPCIONES VECTORES")
    print("1. Suma de vectores")
    print("2. Producto punto")
    print("3. Producto cruz")
    print("4. Producto de Hadamard")
    print("5. Salir")

    opcion = input("Elija una opcion: ")

    if opcion in ['1', '2', '3', '4']:
        v1 = pedir_vector("Ingrese el vector 1 separado por espacios: ")
        v2 = pedir_vector("Ingrese el vector 2 separado por espacios: ")

        if opcion == '1':
            if len(v1) == len(v2):
                resultado = [a + b for a, b in zip(v1, v2)]
                print("Suma:", resultado)
            else:
                print("Error: Los vectores deben tener la misma dimension.")

        elif opcion == '2':
            if len(v1) == len(v2):
                resultado = sum(a * b for a, b in zip(v1, v2))
                print("Producto punto:", resultado)
            else:
                print("Error: Los vectores deben tener la misma dimension.")

        elif opcion == '3':
            if len(v1) == 3 and len(v2) == 3:
                resultado = [
                    v1[1]*v2[2] - v1[2]*v2[1],
                    v1[2]*v2[0] - v1[0]*v2[2],
                    v1[0]*v2[1] - v1[1]*v2[0]
                ]
                print("Producto cruz:", resultado)
            else:
                print("Error: El producto cruz requiere vectores de 3 dimensiones.")

        elif opcion == '4':
            if len(v1) == len(v2):
                resultado = [a * b for a, b in zip(v1, v2)]
                print("Producto de Hadamard:", resultado)
            else:
                print("Error: Los vectores deben tener la misma dimension.")

    elif opcion == '5':
        print("chau profe")
        break
    else:
        print("Opcion no valida.")
```

```python
def ingresar_matriz():
    while True:
        try:
            filas = int(input("Ingrese el numero de filas: "))
            columnas = int(input("Ingrese el numero de columnas: "))
            if filas > 0 and columnas > 0:
                break
            print("Dimensiones deben ser mayores a 0.")
        except ValueError:
            print("Error: Ingrese numeros enteros.")

    matriz = []
    for i in range(filas):
        while True:
            try:
                fila = list(map(float, input(f"Ingrese la fila {i+1} separada por espacios: ").split()))
                if len(fila) != columnas:
                    print(f"Error: debe ingresar exactamente {columnas} valores.")
                else:
                    matriz.append(fila)
                    break
            except ValueError:
                print("Error: Ingrese solo numeros separados por espacios.")
    return matriz

while True:
    print("\nMENU DE OPCIONES MATRICES")
    print("1. Suma de matrices")
    print("2. Multiplicacion de matrices")
    print("3. Producto de Hadamard")
    print("4. Producto de Kronecker")
    print("5. Salir")

    opcion = input("Elija una opcion: ")

    if opcion in ['1', '2', '3', '4']:
        print("--- Matriz A ---")
        A = ingresar_matriz()
        print("--- Matriz B ---")
        B = ingresar_matriz()

        fA, cA = len(A), len(A[0])
        fB, cB = len(B), len(B[0])

        if opcion == '1':
            if fA == fB and cA == cB:
                R = [[A[i][j] + B[i][j] for j in range(cA)] for i in range(fA)]
                for fila in R: print(fila)
            else:
                print("Error: Deben tener las mismas dimensiones.")

        elif opcion == '2':
            if cA == fB:
                R = [[sum(A[i][k] * B[k][j] for k in range(cA)) for j in range(cB)] for i in range(fA)]
                for fila in R: print(fila)
            else:
                print("Error: Columnas de A deben ser iguales a filas de B.")

        elif opcion == '3':
            if fA == fB and cA == cB:
                R = [[A[i][j] * B[i][j] for j in range(cA)] for i in range(fA)]
                for fila in R: print(fila)
            else:
                print("Error: Deben tener las mismas dimensiones.")

        elif opcion == '4':
            R = [[A[i // fB][j // cB] * B[i % fB][j % cB] for j in range(cA * cB)] for i in range(fA * fB)]
            for fila in R: print(fila)

    elif opcion == '5':
        print("se acabo el programa caballeros")
        break
    else:
        print("Opcion no valida.")
```

```python
def ingresar_matriz2():
    while True:
        try:
            filas = int(input("Ingrese el numero de filas: "))
            columnas = int(input("Ingrese el numero de columnas: "))
            if filas > 0 and columnas > 0:
                break
            print("Dimensiones deben ser mayores a 0.")
        except ValueError:
            print("Error: Ingrese numeros enteros.")

    matriz = []
    for i in range(filas):
        while True:
            try:
                fila = list(map(float, input(f"Ingrese la fila {i+1} separada por espacios: ").split()))
                if len(fila) != columnas:
                    print(f"Error: debe ingresar exactamente {columnas} valores.")
                else:
                    matriz.append(fila)
                    break
            except ValueError:
                print("Error: Ingrese solo numeros.")
    return matriz

def determinante(M):
    if len(M) == 1:
        return M[0][0]
    if len(M) == 2:
        return M[0][0]*M[1][1] - M[0][1]*M[1][0]
    det = 0
    for c in range(len(M)):
        menor = [fila[:c] + fila[c+1:] for fila in M[1:]]
        det += ((-1)**c) * M[0][c] * determinante(menor)
    return det

while True:
    print("\nMENU TRANSFORMACIONES")
    print("1. Trasponer matriz")
    print("2. Matriz de cofactores")
    print("3. Vectorizar por filas")
    print("4. Salir")

    opcion = input("Elija una opcion: ")

    if opcion in ['1', '2', '3']:
        A = ingresar_matriz()
        fA, cA = len(A), len(A[0])

        if opcion == '1':
            R = [[A[j][i] for j in range(fA)] for i in range(cA)]
            for fila in R: print(fila)

        elif opcion == '2':
            if fA == cA:
                if fA == 1:
                    print([[1]])
                else:
                    C = []
                    for i in range(fA):
                        fila_C = []
                        for j in range(cA):
                            menor = [fila[:j] + fila[j+1:] for r, fila in enumerate(A) if r != i]
                            cofactor = ((-1)**(i+j)) * determinante(menor)
                            fila_C.append(cofactor)
                        C.append(fila_C)
                    for fila in C: print(fila)
            else:
                print("Error: La matriz debe ser cuadrada.")

        elif opcion == '3':
            vec = [[A[i][j]] for i in range(fA) for j in range(cA)]
            for fila in vec: print(fila)

    elif opcion == '4':
        print("se acabó la tarea profe")
        break
    else:
        print("Opcion no valida.")
```
