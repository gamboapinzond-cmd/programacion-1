```python
mi_lista = ['manzana', 'banana', 'cereza']
print("--- 1. CONVERSIONES DESDE LISTA ---")
print(f"Original (Lista): {mi_lista} | Tipo: {type(mi_lista)}")
print(f"A Tupla: {tuple(mi_lista)} | Tipo: {type(tuple(mi_lista))}")
print(f"A Conjunto: {set(mi_lista)} | Tipo: {type(set(mi_lista))}")
print(f"A Diccionario: {dict(enumerate(mi_lista))} | Tipo: {type(dict(enumerate(mi_lista)))}\n")

mi_tupla = (10, 20, 30)
print("--- 2. CONVERSIONES DESDE TUPLA ---")
print(f"Original (Tupla): {mi_tupla} | Tipo: {type(mi_tupla)}")
print(f"A Lista: {list(mi_tupla)} | Tipo: {type(list(mi_tupla))}")
print(f"A Conjunto: {set(mi_tupla)} | Tipo: {type(set(mi_tupla))}")
print(f"A Diccionario: {dict(enumerate(mi_tupla))} | Tipo: {type(dict(enumerate(mi_tupla)))}\n")

mi_diccionario = {'a': 1, 'b': 2, 'c': 3}
print("--- 3. CONVERSIONES DESDE DICCIONARIO ---")
print(f"Original (Diccionario): {mi_diccionario} | Tipo: {type(mi_diccionario)}")
print(f"A Lista: {list(mi_diccionario.items())} | Tipo: {type(list(mi_diccionario.items()))}")
print(f"A Tupla: {tuple(mi_diccionario.items())} | Tipo: {type(tuple(mi_diccionario.items()))}")
print(f"A Conjunto: {set(mi_diccionario.items())} | Tipo: {type(set(mi_diccionario.items()))}\n")

mi_conjunto = {100, 200, 300}
print("--- 4. CONVERSIONES DESDE CONJUNTO ---")
print(f"Original (Conjunto): {mi_conjunto} | Tipo: {type(mi_conjunto)}")
print(f"A Lista: {list(mi_conjunto)} | Tipo: {type(list(mi_conjunto))}")
print(f"A Tupla: {tuple(mi_conjunto)} | Tipo: {type(tuple(mi_conjunto))}")
print(f"A Diccionario: {dict(enumerate(mi_conjunto))} | Tipo: {type(dict(enumerate(mi_conjunto)))}\n")
```
