```python
import math
math.pi
```

```python
def volumen(r1, r2, h, pi=math.pi):
  return (((h*pi)/3)*(r1**2+r1*r2+r2**2))
```

```python
radio1 = float(input("ingrese el radio 1: "))
radio2 = float(input("ingrese el radio 2: "))
altura = float(input("ingrese la altura: "))

resultado = volumen(radio1, radio2, altura)
print("El volumen es:", resultado)
```

```python
def contador_vocales(palabra):
  vocales = "aeiou"
  consonantes = "bcdfghjklmnñpqrstvwxyz"
  contador = 0
  contador1 = 0
  contador3 = 0

  for letra in palabra.lower():
    if letra in vocales:
      contador += 1
    elif letra in consonantes:
      contador1 += 1
    else:
      contador3 += 1
  print("Cantidad de vocales:", contador)
  print("Cantidad de consonates:", contador1)
  print("Cantidad de caracteres distintos a vocales y consonantes:", contador3)
```

```python
contador_vocales(palabra="hola 45((&")
```

```python
pal = str(input("ingrese una palabra: "))
contador_vocales(pal)+-
```
