```python
for n in range(0, 10):
  if n == 1 or n == 0:
    print("El número es ", n)
  else:
    for x in range(2, n):
      if n % x == 0:
          print(n, 'es igual a', x, '*', n // x)
          break
    else: # Este else corresponde al for, no al if
      print(n, 'es un número primo')
```

```python
clave_correcta = "python123"
intento = ""
intentos_maximos = 3
intentos_realizados = 0

while intento != clave_correcta and intentos_realizados < intentos_maximos:
    intento = input(f"Ingresa la contraseña (Intento {intentos_realizados + 1}/{intentos_maximos}): ")
    intentos_realizados += 1

if intento == clave_correcta:
    print("¡Acceso concedido!")
else:
    print("Número máximo de intentos alcanzado. Acceso denegado.")
```

```python
clave_correcta = "python123"
intentos = 3
while intentos > 0:
    intentos = input("Ingresa la contraseña: ")
    if intentos == clave_correcta:
      print("¡Acceso concedido!")
      break
    else:
        intentos -= 1
        print("intentos restantes:", intentos)
else:
        print("Número máximo de intentos alcanzado. Acceso denegado.")
```
